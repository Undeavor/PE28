On souhaite optimiser un portefeuille en ayant la prdiction des rendements et des volatilits d'un ensemble d'actifs  n lments.

On a $\sum_iw_i=1$.

Pour tablir la volatilit d'un portefeuille de rendement $R_P=\sum_iw_iR_i$, il est ncessaire de connaitre les covariances entre les diffrents rendements et donc de l'estimer  travers d'un modle choisi.


# Deux choix d'optimisation sont possibles :    

## 1 - Maximisation du Sharpe
Soit $w = (w_1,  \ldots, w_n)^T$ l'ensemble des poids associs aux actifs,
$R_p = (R_1, \ldots, R_n)^{T}$ le vecteur des rendements esprs,
$\Sigma$ la matrice de covariance des rendements, et $r_f$ le taux sans risque.

Alors le ratio de Sharpe du portefeuille s'crit : ( les termes entre parenthses sont les notations matricielles )
$$
\text{Sharpe} = \frac{\sum_iw_iR_i - R_f}{\sigma_p}
= \frac{\sum_iw_i(R_i - R_f)}{\sqrt{\sum_i w_i^2\sigma_i^2+\sum_{ij}w_iw_j\text{cov}(r_i,r_j)}}(= \frac{w^{T}(R_p - R_f \mathbf{1})}{\sqrt{w^{T} \Sigma w}})
$$
o
$\sum_iw_iR_p - R_f$ est le rendement excdentaire espr et
$\sigma_p = \sqrt{w^{T} \Sigma w}$ reprsente la volatilit du portefeuille.

Lobjectif consiste  maximiser ce ratio sous la contrainte de normalisation :
$$
\max_{w} \frac{\sum_iw_i(R_i - R_f)}{\sqrt{\sum_i w_i^2\sigma_i^2+\sum_{ij}w_iw_j\text{cov}(r_i,r_j)}}
\quad \text{avec} \quad \sum_i w_i= 1(=\mathbf{1}^{T}w )
$$

---

## 2 - Minimisation de la volatilit pour un rendement cible

On cherche ici  obtenir le portefeuille le moins risqu possible
pour un rendement espr fix $R_{t}$ :
$$\min_{w} \; \sum_i w_i^2\sigma_i^2+\sum_{ij}w_iw_j\text{cov}(r_i,r_j) \quad \text{avec} \quad \begin{cases} \sum_i w_i= 1(=\mathbf{1}^{T}w ), \\\ \sum_iw_iR_i = R_{t}(=R_p^{T}w).\end{cases}$$

---

Cette formulation permet de tracer la frontire d'efficience,
ensemble des portefeuilles offrant le rendement maximal pour un niveau de risque donn
ou, inversement, le risque minimal pour un rendement donn.

### Hypothses pour simplifier le modle dans un premier temps :
- les titres sont scindables  souhait
- pas de frais
- pas de dividendes
- pas de contrainte de signe pour les poids ( long/short autoriss )

### Dmonstration de l'optimisation du sharpe dans l'hypothse de corrlations nulles:
ici on ne diversifie pas, $ \sqrt{w^{T} \Sigma w}=\sqrt{\sum_i w_i^2\sigma_i^2}$

$$\text{Sharpe}=\frac{\sum_iw_i(R_i - R_f)}{\sqrt{\sum_i w_i^2\sigma_i^2}}\text{et} \sum_i w_i= 1$$
On remarque que le sharpe est invariant par homothtie positive, on prend $\sqrt{\sum_i w_i'^2\sigma_i^2}=1\frac{1}{2}\sum_i w_i'^2\sigma_i^2=\frac{1}{2}$.
 On applique le thorme de Lagrange,
 $$\Delta\text{Sharpe}+\frac{\lambda}{2}\Delta(\sum_i w_i'^2\sigma_i^2-1)=0 \\  \forall i \in [1,n],\ (R_i-R_f)+\lambda(w_i'\sigma_i^2)=0\\  \forall i \in [1,n],\ w_i'=\frac{(R_f-R_i)}{\lambda\sigma_i^2}$$
 Or l'on souhaite aussi $\sum_i w_i= 1$ donc l'on normalise le vecteur:
 $$w_i=\frac{w_i'}{\sum_k w_k'}$$ ( ne dpend pas de $\lambda$ )


### Dmonstration de l'optimisation du sharpe dans le cas gnral:

$$\text{Sharpe}=\frac{\sum_iw_i(R_i - R_f)}{\sqrt{\sum_i w_i^2\sigma_i^2+\sum_{ij}w_iw_j\text{cov}(r_i,r_j)}}\text{et} \sum_i w_i= 1$$
On remarque que le sharpe est invariant par homothtie positive, on prend $\sqrt{\sum_i w_i'^2\sigma_i^2+\sum_{ij}w_i'w_j'\text{cov}(r_i,r_j)}=1\frac{1}{2}(\sum_i w_i'^2\sigma_i^2+\sum_{ij}w_i'w_j'\text{cov}(r_i,r_j))=\frac{1}{2}$.
On applique le thorme de Lagrange,
$$\Delta\text{Sharpe}+\frac{\lambda}{2}\Delta(\sum_i w_i'^2\sigma_i^2+\sum_{ij}w_i'w_j'\text{cov}(r_i,r_j)-1)=0
\\  \forall i \in [1,n],\ (R_p-R_f)+\lambda(w_i'\sigma_i^2+\sum_{ij}w_j'\frac{\text{cov}(r_i,r_j)}{2})=0
$$
En prenant $L_i-L_j$ pour $ij$ :
$$
w_i'(\sigma_i^2-\frac{cov(r_i,r_j)}{2})=w_j'(\sigma_j^2-\frac{cov(r_i,r_j)}{2})$$
D'o,
$$w_i'=\frac{R_f-R_p}{\lambda(\sigma_i^2+\sum_{ij}(\frac{(\sigma_i^2-\frac{cov(r_i,r_j)}{2})}{(\sigma_j^2-\frac{cov(r_i,r_j)}{2})}\frac{cov(r_i,r_j)}{2}))}$$
Or l'on souhaite aussi $\sum_i w_i= 1$ donc l'on normalise le vecteur:
$$w_i=\frac{w_i'}{\sum_k w_k'}$$
( ne dpend pas de $\lambda$ )

Pas encore montr:
- ( opti sharpe ) La solution analytique (en absence de contraintes de signe sur les poids) est donne par :
$$
w^{*} = \frac{\Sigma^{-1}(\mu - r_f \mathbf{1})}
{\mathbf{1}^{T}\Sigma^{-1}(\mu - r_f \mathbf{1})}
$$
- ( opti volatilit avec rendement fixe ) En notant :
$$A = \mathbf{1}^{T}\Sigma^{-1}\mathbf{1}, \quad B = \mathbf{1}^{T}\Sigma^{-1}\mu, \quad C = \mu^{T}\Sigma^{-1}\mu, \quad D = AC - B^{2},$$
la solution optimale scrit :
$$w(\mu_t) = \Sigma^{-1}\left[ \frac{C - B\mu_t}{D}\mathbf{1} + \frac{A\mu_t - B}{D}\mu\right]$$

  La variance minimale associe  ce portefeuille est alors :
$$\sigma_p^{2}(\mu_t) = \frac{A\mu_t^{2} - 2B\mu_t + C}{D}.$$

```python
import numpy as np

def optimal_sharpe_weights_vectorized(returns, cov_matrix, rf=0):
    """
    Calcule les poids du portefeuille qui maximisent le Sharpe ratio
    dans le cas gnral, de manire vectorise.
    Paramtres :
    -----------
    returns : array_like
        Rendements attendus des actifs (R_i)
    cov_matrix : array_like
        Matrice de covariance des rendements (_i^2 et cov(r_i,r_j))
    rf : float
        Taux sans risque
    Retour :
    -------
    weights : np.array
        Poids optimaux normaliss
    """
    returns = np.array(returns)
    cov_matrix = np.array(cov_matrix)

    # Matrice pour le calcul vectoris de sum_term
    sigma_diag = np.diag(cov_matrix)
    n = len(returns)

    # Calcul des termes numerators / denominators pour tous i,j avec i != j
    cov_matrix_half = cov_matrix / 2
    numerator = sigma_diag[:, None] - cov_matrix_half
    denominator = sigma_diag - cov_matrix_half.T

    # viter division par zro sur la diagonale
    np.fill_diagonal(denominator, 1)  # nimporte quelle valeur non nulle

    # somme des termes hors diagonale
    sum_term = np.sum(numerator / denominator * cov_matrix_half, axis=1) - np.diag(numerator / denominator * cov_matrix_half)

    # w' vectoris
    w_prime = (rf - np.mean(returns)) / (sigma_diag + sum_term)

    # Normalisation
    weights = w_prime / np.sum(w_prime)

    return weights

# Exemple d'utilisation
returns = [0.12, 0.10, 0.08]
cov_matrix = [
    [0.04, 0.006, 0.004],
    [0.006, 0.03, 0.002],
    [0.004, 0.002, 0.02]
]
rf = 0.03

weights = optimal_sharpe_weights_vectorized(returns, cov_matrix, rf)
print("Poids optimaux vectoriss :", weights)
```

##Optimisation de A,B,C par MLL: ( tout est toujours conditionnel ici )

On pose $R(t)=(r_1(t),\ldots,r_n(t))^T \sim N(0,H(t))$ le vecteur gaussien des rendements de densit :

$$L(R) = \frac{1}{(2\pi)^{N/2} |H|^{1/2}}
\exp\!\Big(-\frac{1}{2}< H^{-1} (R - 0), (R - 0) >\Big)$$
On prend le produit scalaire de vecteurs usuel $<M,N>=M^TN$
$$L(R) = \frac{1}{(2\pi)^{N/2} |H|^{1/2}}
\exp\!\Big(-\frac{1}{2}R^T H^{-1} R\Big)$$
D'o l'expression de la vraisemblance avec Bayes:
$$L(R(1), \ldots,R(T))=\prod_{i=1}^T \frac{1}{(2\pi)^{N/2} |H(i)|^{1/2}}
\exp\!\Big(-\frac{1}{2}R^T(i) H^{-1}(i) R(i)\Big)$$
$$\boxed{LL(R(1), \ldots,R(T))=\sum_{i=1}^T -\frac{N}{2}log(2\pi)-\frac{1}{2}log|H(i)|
-\frac{1}{2}R^T(i) H^{-1}(i) R(i)}$$
  
-On dfinit le produit scalaire matriciel comme tant:
$$<A,B>=Tr(A^TB)$$

En scalaire, on a:
$$df = \left(\frac{\partial f}{\partial X}\right)^\top dX$$

Donc les diffrencielles matricielles deviennent:
$$df = \operatorname{tr}\!\left(\left(\frac{\partial f}{\partial X}\right)^\top dX\right)$$

-De plus,
$$(H+dH)(H^{-1}+dH^{-1})=I$$
$$ HH^{-1}+HdH^{-1}+dHH^{-1}+dHdH^{-1}=I$$
$$ HdH^{-1}+dHH^{-1}+dHdH^{-1}=0$$
$$\text{ avec } dHdH^{-1} \text{ ngligeable devant les autres termes}$$
$$ HdH^{-1}=-dHH^{-1}$$
$$ dH^{-1}=-H^{-1}dHH^{-1}$$

-Encore, si les valeurs propres de H sont les $\lambda_i$
$$d(log(|H|))=d(\log |H|) = d\left(\sum_i \log \lambda_i \right) = \sum_i \frac{d\lambda_i}{\lambda_i} = \operatorname{Tr}(H^{-1} dH)$$


-De surcrot, soit A vecteur, B matrice
$$A^TB^{-1}dB B^{-1} A=\operatorname{Tr}(A^TB^{-1}dB B^{-1} A)$$ Or $$\operatorname{Tr}(AB)=\operatorname{Tr}(BA)$$ Donc $$A^TB^{-1}dB B^{-1} A=\operatorname{Tr}(B^{-1} AA^TB^{-1}dB)$$
-On a finalement,
$$dLL=-\frac{1}{2}d(log(|H(i)|))-\frac{1}{2}R^T(i)(dH(i)^{-1}) R(i)$$
$$dLL= -\frac{1}{2}\operatorname{Tr}(H(i)^{-1} dH(i))-\frac{1}{2}R^T(i)(-H(i)^{-1}dH(i)H(i)^{-1}) R(i)$$
$$dLL=-\frac{1}{2}\operatorname{Tr}(H(i)^{-1} dH(i))+\frac{1}{2}\operatorname{Tr}(H(i)^{-1} R(i)R(i)^TH(i)^{-1}dH(i))$$
$$\boxed{\frac{\partial LL}{\partial H(i)}=-\frac{1}{2}H(i)^{-1}+\frac{1}{2}H(i)^{-1} R(i)R(i)^TH(i)^{-1}}$$
On peut donc calculer les drives matricielles par rapport  A, B et C grce  la rgle de la chaine: $$\frac{\partial LL}{\partial A}=\sum_i\frac{\partial LL}{\partial H(i)}\frac{\partial H(i)}{\partial A}$$
donc effectuer une descente de gradient.

```python
%matplotlib inline
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# ----------------------------
# 1 Fonctions de calcul
# ----------------------------
def safe_sharpe(w, R, Sigma):
    ret = np.dot(w, R)
    vol2 = np.dot(w.T, np.dot(Sigma, w))
    if vol2 <= 0:  # viter racine ngative
        return 1e6
    vol = np.sqrt(vol2)
    return -ret / vol  # minimisation

def portfolio_vol(weights, Sigma):
    vol2 = np.dot(weights.T, np.dot(Sigma, weights))
    return np.sqrt(vol2) if vol2 > 0 else 1e6

def min_vol_target(Rt, R, Sigma, w0, bounds):
    cons = (
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
        {'type': 'eq', 'fun': lambda w: np.dot(w, R) - Rt}
    )
    res = minimize(portfolio_vol, w0, args=(Sigma,), bounds=bounds, constraints=cons)
    return res.fun, res.x

def optimize_sharpe(R, Sigma, w0, bounds):
    cons = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
    res = minimize(safe_sharpe, w0, args=(R, Sigma), bounds=bounds, constraints=cons)
    w_sharpe = res.x
    vol_sharpe = np.sqrt(np.dot(w_sharpe.T, np.dot(Sigma, w_sharpe)))
    ret_sharpe = np.dot(w_sharpe, R)
    sharpe_max = ret_sharpe / vol_sharpe
    return sharpe_max, w_sharpe, vol_sharpe, ret_sharpe

# ----------------------------
# 2 Fonction interactive
# ----------------------------
def interactive_portfolio(r1=0.12, r2=0.10, r3=0.15, r4=0.09, Rt_target=0.12):
    R = np.array([r1, r2, r3, r4])
    Sigma = np.array([
        [0.1**2, 0.01, 0.02, 0.01],
        [0.01, 0.08**2, 0.01, 0.005],
        [0.02, 0.01, 0.12**2, 0.015],
        [0.01, 0.005, 0.015, 0.09**2]
    ])
    n = len(R)
    w0 = np.ones(n)/n
    # Limiter shorts pour viter nan
    bounds = [(-0.1, 1) for _ in range(4)]

    # Sharpe max
    sharpe_max, w_sharpe, vol_sharpe, ret_sharpe = optimize_sharpe(R, Sigma, w0, bounds)

    # Volatilit min pour Rt_target
    vol_min, w_min_vol = min_vol_target(Rt_target, R, Sigma, w0, bounds)

    # Frontire efficiente
    Rt_range = np.linspace(min(R), max(R), 50)
    vols = [min_vol_target(Rt_val, R, Sigma, w0, bounds)[0] for Rt_val in Rt_range]

    # Plot
    plt.figure(figsize=(8,5))
    plt.plot(vols, Rt_range, label='Frontire efficiente', color='blue')
    plt.scatter(vol_sharpe, ret_sharpe, color='red', label='Sharpe max', zorder=5)
    plt.scatter(vol_min, Rt_target, color='green', label='Min vol portefeuille', zorder=5)
    plt.xlabel('Volatilit')
    plt.ylabel('Rendement')
    plt.title('Frontire efficiente interactive')
    plt.legend()
    plt.grid(True)
    plt.show()

    print("Poids Sharpe max :", np.round(w_sharpe,4))
    print("Rendement Sharpe max :", np.round(ret_sharpe,4))
    print("Volatilit Sharpe max :", np.round(vol_sharpe,4))
    print(f"Poids Min vol (Rt={Rt_target}):", np.round(w_min_vol,4))
    print("Volatilit min :", np.round(vol_min,4))

# ----------------------------
# 3 Widgets interactifs
# ----------------------------
style = {'description_width': '80px'}
r_sliders = [
    widgets.FloatSlider(value=val, min=0.0, max=0.25, step=0.01, description=f'R{i+1}', style=style)
    for i, val in enumerate([0.12, 0.10, 0.15, 0.09])
]
Rt_slider = widgets.FloatSlider(value=0.12, min=0.05, max=0.20, step=0.005, description='Rt cible', style=style)

ui = widgets.VBox(r_sliders + [Rt_slider])

out = widgets.interactive_output(
    interactive_portfolio,
    {'r1': r_sliders[0], 'r2': r_sliders[1], 'r3': r_sliders[2], 'r4': r_sliders[3],
     'Rt_target': Rt_slider}
)

display(ui, out)
```

Annexe : Dmonstration de la densit de la loi gaussienne multivarie

Soit $X \in \mathbb{R}^n$ un vecteur alatoire gaussien d'esprance
$\mu \in \mathbb{R}^n$ et de matrice de covariance
$\Sigma \in \mathbb{R}^{n \times n}$, symtrique dfinie positive.

On cherche  dmontrer que la densit de $X$ est donne par :

$$
f_X(x) = \frac{1}{(2\pi)^{n/2} |\Sigma|^{1/2}}
\exp\Big(-\tfrac{1}{2}(x-\mu)^\top \Sigma^{-1}(x-\mu)\Big),
\qquad x \in \mathbb{R}^n.
$$



1. Cas standard

Considrons d'abord un vecteur alatoire standard
$Z \sim \mathcal{N}(0, I_n)$.
Sa densit est :

$$
f_Z(z) = (2\pi)^{-n/2} \exp\Big(-\tfrac{1}{2} z^\top z\Big),
\qquad z \in \mathbb{R}^n.
$$



2. Transformation linaire

Puisque $\Sigma$ est symtrique dfinie positive,
il existe une matrice inversible L telle que :

$$
\Sigma = L L^\top.
$$

(On peut choisir L comme la matrice de Cholesky de $\Sigma$.)

Dfinissons maintenant :

$$
X = \mu + LZ.
$$

Alors :

$$
\mathbb{E}[X] = \mu, \qquad
\text{Cov}(X) = L , \text{Cov}(Z) , L^\top = L I_n L^\top = \Sigma.
$$

Ainsi, $X$ est bien un vecteur gaussien de moyenne $\mu$ et de covariance $\Sigma$.



3. Changement de variables

On a le changement de variables :

$$
z = L^{-1}(x - \mu),
$$

dont le jacobien est :

$$
|\det(L^{-1})| = |\det L|^{-1}.
$$

La densit de $X$ est donc :

$$
f_X(x) = f_Z(L^{-1}(x - \mu)) , |\det(L^{-1})|.
$$



4. Substitution et simplification

En remplaant $f_Z$ :

$$
f_X(x) = (2\pi)^{-n/2}
\exp\Big(-\tfrac{1}{2}(L^{-1}(x-\mu))^\top (L^{-1}(x-\mu))\Big)
, |\det(L^{-1})|.
$$

Or :

$$
(L^{-1}(x-\mu))^\top (L^{-1}(x-\mu))
= (x-\mu)^\top (L^{-1})^\top L^{-1} (x-\mu)
= (x-\mu)^\top \Sigma^{-1} (x-\mu).
$$

De plus :

$$
|\det(L^{-1})| = |\det L|^{-1} = |\Sigma|^{-1/2}.
$$



5. Forme finale

En regroupant les termes, on obtient :

$$
\boxed{
f_X(x) = \frac{1}{(2\pi)^{n/2} |\Sigma|^{1/2}}
\exp\Big(-\tfrac{1}{2}(x-\mu)^\top \Sigma^{-1}(x-\mu)\Big)
}.
$$



Remarque

Si $\Sigma$ est seulement semi-dfinie positive (et non inversible),
la densit n'existe pas au sens usuel sur $\mathbb{R}^n$ :
la loi est alors concentre sur un sous-espace affine de dimension gale au rang de $\Sigma$.
